```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv')
df.head()
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
df.head()
```

```python
location_counts = df.location.value_counts()
location_counts
```

```python
others_locations = location_counts[location_counts <= 10]
others_locations
```

```python
df.location = df['location'].apply(lambda x: 'other' if x in others_locations else x)
df.head()
```
